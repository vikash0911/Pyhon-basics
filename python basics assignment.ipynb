**2.** What is an interpreter in Python?
**ans-** An interpreter in Python is a program that reads and executes Python code line by line. Unlike compiled languages (where the source code is translated into machine code before execution), an interpreter directly executes the instructions in the source code, translating them into machine-readable code in real-time.

**5.** What is mutability in Python?
**ans-** Mutability in Python refers to the ability of an object to be changed after it is created. Some objects in Python are mutable, meaning their state or contents can be modified, while others are immutable, meaning once they are created, they cannot be altered.

**6.** Why are lists mutable, but tuples are immutable?

**7.** What is the difference between “==” and “is” operators in Python?
**ans-** The == operator in Python checks if two objects have equal values (i.e., their contents are the same), whereas the is operator checks if two variables point to the same object in memory (i.e., they are the same object).

**8. **What are logical operators in Python?
**ans- **In Python, logical operators are used to perform logical operations on boolean values (True or False). These operators are commonly used to combine conditional statements and evaluate expressions.

The Three Logical Operators in Python:
(a). and: Returns True if both operands (conditions) are True; otherwise, returns False.
(b). or: Returns True if at least one of the operands (conditions) is True; if both are False, it returns False.

**9.** What is type casting in Python?
**ans-** Type casting in Python refers to the process of converting one data type into another. This is useful when you need to change the type of a value, for example, converting a string that represents a number into an integer or floating-point number, or converting a number to a string for concatenation.

**10.** What is the difference between implicit and explicit type casting?
**ans-**  (a). Implicit Type Casting (Automatic Conversion):
What it is: Implicit type casting (also called type coercion) is when Python automatically converts one data type to another without any explicit instruction from the programmer.
When it happens: This happens when Python converts a lower data type to a higher data type, usually without loss of information (e.g., from int to float).
Why it happens: It occurs when the operation requires it, and Python handles the conversion for you to avoid errors.

(b). Explicit Type Casting (Manual Conversion):
What it is: Explicit type casting requires the programmer to manually convert one data type to another using type casting functions like int(), float(), str(), etc.
When it happens: It happens when you explicitly tell Python to convert the type.
Why it happens: You may need to convert a value to a specific type to perform an operation, ensure compatibility, or avoid errors.

**12. **How does the elif statement work?
**ans-** The elif statement in Python is short for "else if" and is used to check multiple conditions in a sequence. It allows you to test several conditions in a structured way, and if the first condition (if) is false, it moves on to check the elif conditions. If an elif condition is true, the corresponding block of code is executed. If none of the if or elif conditions are true, the else block (if present) will be executed.

How elif works:
(a). The if statement is checked first. If it's True, the associated code block runs, and the rest of the conditions are ignored.
(b).If the if statement is False, the program moves to the first elif statement. If the elif condition is True, the corresponding block of code runs, and the remaining conditions are skipped.
(c). If the elif condition is also False, the program checks the next elif (if any).
(d). If none of the conditions (if or elif) are true, and there is an else statement, the else block is executed.

**13.** What is the difference between for and while loops?
**ans-** The for and while loops are both used to repeat a block of code multiple times, but they differ in how they control the repetition. Here’s a breakdown of the key differences:

(a). for loop:
Purpose: The for loop is used when you know exactly how many times you want to iterate over a sequence (such as a list, tuple, string, or range) or a specific number of iterations.
Control: The for loop iterates over a collection (like a list or range) or a sequence and automatically stops when it reaches the end of the collection.

(b). while loop:
Purpose: The while loop is used when you want to repeat a block of code while a certain condition is True. It continues until the condition becomes False.
Control: The while loop depends on a condition that is checked before each iteration. If the condition is True, the loop continues; if the condition becomes False, the loop stops.

Here’s why a while loop is more suitable than a for loop in this case:

(a). Unknown Number of Iterations: The program doesn't know in advance how many times the user will need to try to enter a valid email. The number of attempts could vary.

(b). Condition-Based Loop: The loop should continue running as long as the condition (the email being valid) is not met. Once the user enters a valid email, the loop should stop.

(c). Dynamic Termination: The loop is controlled by the condition (whether the email is valid or not), and it should stop only when the condition becomes true. A while loop is perfect for this scenario because it can repeat indefinitely until the user provides the correct input.
